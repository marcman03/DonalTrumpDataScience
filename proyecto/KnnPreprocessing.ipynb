# KNN PREPROCESSING

```python
%matplotlib inline
import pandas
import matplotlib.pyplot as plt
import numpy as np
X = pandas.read_csv('X.csv', sep=',', na_values="")
y = pandas.read_csv('y.csv', sep=',', na_values="")
print(X.head())
print(y.head())
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
import sklearn.neighbors as nb

# Asegurarse de que y es un array unidimensional
y = y.values.ravel()

# Configuración de la validación cruzada
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=1)

cv_scores = cross_val_score(nb.KNeighborsClassifier(), X=X,y=y, cv=cv, scoring='accuracy')  
np.mean(cv_scores)
```

```python
Xaux = X.to_numpy()
for i in range(20):
    print('std:',Xaux[:,i].std(),'min',Xaux[:,i].min(),'max',Xaux[:,i].max())
```

```python
from sklearn import preprocessing

# One way is to standarize all data mean 0, std 1
scaler = preprocessing.StandardScaler().fit(X)
X2=scaler.transform(X)

for i in range(20):
    print('std:',X2[:,i].std(),'min',X2[:,i].min(),'max',X2[:,i].max())
```

```python
cv_scores = cross_val_score(nb.KNeighborsClassifier(), X=X2,y=y, cv=cv, scoring='accuracy')  
np.mean(cv_scores)
```
